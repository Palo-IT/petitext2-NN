```python
import matplotlib.pyplot as plt
import numpy as np 

process_time = np.array([])

em_score = np.array([69.97161778618732, 69.8864711447493, 70.7663197729423, 70.52034058656575])
f1_score = np.array([78.7595992310913, 78.65561683643998, 79.32997570934039, 79.10542150374778])
word_emb_type = ('glove','glove','fasttext','fasttext')


ax =  fig = plt.figure(figsize=(20, 8)) 
plt.suptitle('R-NET : EM and F1 score with glove and fasttext word embedding',fontsize=16)

plt.subplot(121)
y_pos = np.arange(len(word_emb_type))
plt.plot(y_pos, em_score)
plt.xticks(y_pos, word_emb_type)
plt.ylabel('EM')
plt.title('EM score')
 

plt.subplot(122)
y_pos = np.arange(len(word_emb_type))
plt.plot(y_pos, f1_score)
plt.xticks(y_pos, word_emb_type)
plt.ylabel('F1')
plt.title('F1 score')
 
plt.show()
```
